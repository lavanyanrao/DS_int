```python
# IMDB top 100

def f2(url1,url2):
    
    Name = []
    Rating=[]
    Year=[]
    
    for i in (url1,url2):
        page = requests.get(i)
        soup=BeautifulSoup(page.content,'html.parser')

        movies = soup.find_all('div',class_='lister-item-content')

        for i in movies:
            name = i.h3.a.text
            Name.append(name)
            year = i.find('span',class_='lister-item-year text-muted unbold').text[1:5]
            Year.append(year)
            rating = i.strong.text
            Rating.append(rating)
            
        global IMDB_list
        IMDB_list = pd.DataFrame({'Name':Name,'Year':Year,'Rating':Rating})
        
    return IMDB_list
```

```python
f2('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc','https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt')
```

```python
# IMDB top 100 Indian movies

def f3(url): 
    page = requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')

    Name = []
    Rating=[]
    Year=[]
    
    movies = soup.find_all('td',class_='titleColumn')
    
    for i in movies[0:100]:
        name = i.a.text
        Name.append(name)
        year = i.find('span',class_='secondaryInfo').text[1:5]
        Year.append(year)
        
    m2 = soup.find_all('td',class_='ratingColumn imdbRating')
    
    for i in m2[0:100]:
        rating = i.strong.text
        Rating.append(rating)
    
    
    global IMDB_top100_Indian
    IMDB_top100_Indian = pd.DataFrame({'Name':Name,'Year':Year,'Rating':Rating})
    return IMDB_top100_Indian
```

```python
f3('https://www.imdb.com/india/top-rated-indian-movies/')
```

```python
IMDB_top100_Indian.to_csv("IMDB_top100_Indian.csv")
```

```python
# Details and reviews of any 5 books 

def f4(url1,url2,url3,url4,url5):
    
    Name = []
    Author = []
    Genre = []
    Review = []
    
    for i in [url1,url2,url3,url4,url5]:
        
        page = requests.get(i)
        soup = BeautifulSoup(page.content,'html.parser')
    
        content = soup.find_all('div',class_='book-details')
        name=content[0].h4.text[1:-1]
        Name.append(name)
        author=content[0].find('p',class_='sans bold').a.text
        Author.append(author)
        genre = content[0].find('p',class_='genre-links').text[1:-1]
        Genre.append(genre)
    
        data = soup.find_all('div',class_='article-body')
        review = data[0].text[2:]
        Review.append(review)
        
    Book_info = pd.DataFrame({'Name':Name,'Aurthor':Author,'Genre':Genre,'Review':Review})
    return Book_info
```

```python
# Team rankings

def f5(url):
    page = requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')

    Country = []
    Matches =[]
    Points =[]
    Rating = []

    banner= soup.find_all('tr',class_='rankings-block__banner')

    Country.append(banner[0].find('span',class_='u-hide-phablet').text)
    Matches.append(banner[0].find('td',class_='rankings-block__banner--matches').text)
    Points.append(banner[0].find('td',class_='rankings-block__banner--points').text)
    Rating.append(banner[0].find('td',class_='rankings-block__banner--rating u-text-right').text)

    table_body = soup.find_all('tr',class_='table-body')

    for i in table_body[0:9]:
        Country.append(i.find('span',class_='u-hide-phablet').text)
        Matches.append(i.find('td',class_='table-body__cell u-center-text').text)
        Points.append( i.find_all('td',class_='table-body__cell u-center-text')[1].text)
        Rating.append(i.find('td',class_='table-body__cell u-text-right rating').text)
    
    ODI_team= pd.DataFrame({'Country':Country,'Matches':Matches,'Points':Points,'Rating':Rating})
    return ODI_team
```

```python
# Player rankings

def f6(url):
    page = requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')

    Name = []
    Team =[]
    Rating = []

    banner= soup.find_all('tr',class_='rankings-block__banner')

    Name.append(banner[0].find('div',class_='rankings-block__banner--name-large').text)
    Team.append(banner[0].find('div',class_='rankings-block__banner--nationality').text)
    Rating.append(soup.find_all('td',class_='u-text-left')[0].text)

    table_body = soup.find_all('tr',class_='table-body')

    for i in table_body[0:9]:
        Name.append(i.a.text)
        Team.append(i.find('span',class_='table-body__logo-text').text)
        Rating.append(i.find('td',class_='table-body__cell rating').text)
    
    ODI_player_ranking=pd.DataFrame({'Name':Name,'Team':Team,'Rating':Rating})
    return ODI_player_ranking
   
```

```python
# San Fransisco weather forecast for 7 days

def f7(url):
    
    page = requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    
    Day = []
    Description = []
    
    forecast_odd = soup.find_all('div',class_='row row-odd row-forecast')
    forecast_even = soup.find_all('div',class_='row row-even row-forecast')
    
    for i in range(0,7):
        Day.append(forecast_odd[i].find('div',class_='col-sm-2 forecast-label').text)
        Description.append(forecast_odd[i].find('div',class_='col-sm-10 forecast-text').text)
    
        Day.append(forecast_even[i].find('div',class_='col-sm-2 forecast-label').text)
        Description.append(forecast_even[i].find('div',class_='col-sm-10 forecast-text').text)

    Weather_forecast=pd.DataFrame({'Day':Day,'Description':Description})
    return Weather_forecast    
```

```python
# Software developer jobs

def f8(start_pg_no,end_page_no):
    
    Job_title=[]
    Company = []
    Experience=[]
    Package=[]
    Location=[]
    
    for i in range(start_pg_no,end_page_no):
        page = requests.get(f'https://www.monsterindia.com/search/software-developer-jobs-in-india-{i}?searchId=09a376d1-b18c-42b0-ad9b-34916251846b', headers = {'User-Agent':'Edg/88.0.705.56'})
        soup = BeautifulSoup(page.content,'html.parser')
        jobs = soup.find_all('div',class_='job-tittle')
    
        for i in jobs:
            Job_title.append(i.h3.a.text)
            
            if i.find('span',class_='company-name').a==None:
                Company.append('NaN')
            else:
                Company.append(i.find('span',class_='company-name').a.text)
                
            Location.append(i.find('div',class_='col-xxs-12 col-sm-5 text-ellipsis').text.strip())
            Experience.append(i.find('div',class_='exp col-xxs-12 col-sm-3 text-ellipsis').text.strip())
            Package.append(i.find('div',class_='package col-xxs-12 col-sm-4 text-ellipsis').text.strip())
            
        Jobs= pd.DataFrame({'Job_totle':Job_title,'Company':Company,'Experience':Experience,'Location':Location,'Package':Package})
        
    return Jobs
```

```python
# Amazon.in mobiles under 20,000

def f10(start_pg_no,end_page_no):
    
    Name=[]
    Image_url=[]
    Price=[]
    Rating=[]
        
    for i in range(start_pg_no,end_page_no):
        page = requests.get(f'https://www.amazon.in/s?i=electronics&bbn=1389432031&rh=n%3A1389432031%2Cp_36%3A-2000000&dc&qid=1612541252&rnid=1318502031&ref=sr_pg_{i}',headers = {'User-Agent':'Edg/88.0.705.56'})
        soup=BeautifulSoup(page.content,'html.parser')
        
        mobile=soup.find_all('div',class_='a-section a-spacing-medium')
        
        for j in mobile:
            Name.append(j.find('span',class_='a-size-medium a-color-base a-text-normal').text)
            Image_url.append(j.find('div',class_='a-section aok-relative s-image-fixed-height').img.get('src'))
            Price.append(j.find('span',class_='a-price-whole').text)
            Rating.append(j.find('span',class_='a-icon-alt').text)
            
        Mobiles = pd.DataFrame({'Name':Name,'Image_url':Image_url,'Price':Price,'Rating':Rating})
        
    return Mobiles       
        
```
