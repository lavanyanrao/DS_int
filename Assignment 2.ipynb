```python
# opening the automated window and the desired webpage
driver = webdriver.Chrome("chromedriver.exe") 
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace") 
```

```python
Rating=[]
Review_summary=[]
Full_review=[]
```

```python
# clicking on the all reviews button to show all reviews
all_reviews_btn = driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']")
all_reviews_btn.click()
```

```python
# scraping data off first page of reviews (10 entries)

rating_tag = driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]")
for i in rating_tag:
    Rating.append(i.text[0])
    
review_tag = driver.find_elements_by_xpath("//p[@class='_2-N8zT']")
for i in review_tag:
    Review_summary.append(i.text)
    
freview = driver.find_elements_by_xpath("//div[@class='t-ZTKy']")
for i in freview:
    Full_review.append(i.text)
```

```python
# clicking the next button

nxt_btn = driver.find_elements_by_xpath("//a[@class='_1LKTO3']/span")
nxt_btn[0].click()
driver.implicitly_wait(10)
```

```python
while len(Review_summary)<91:
    
    rating_tag = driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]")
    for i in rating_tag:
        Rating.append(i.text[0])
  
    review_tag = driver.find_elements_by_xpath("//p[@class='_2-N8zT']")
    for i in review_tag:
        Review_summary.append(i.text)

    freview = driver.find_elements_by_xpath("//div[@class='t-ZTKy']")
    for i in freview:
        Full_review.append(i.text)
    
    nxt_btn = driver.find_elements_by_xpath("//a[@class='_1LKTO3']/span")
    nxt_btn[1].click()
    driver.refresh()
    driver.implicitly_wait(10)
```

```python
d6 = pd.DataFrame({'Rating':Rating,'Review_summary':Review_summary,'Full_review':Full_review})
d6
```

```python
driver.quit()
```
